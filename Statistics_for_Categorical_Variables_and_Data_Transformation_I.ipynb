```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
#load and read the dataset
data=pd.read_csv("/content/Titanic Dataset.csv")
data.head()
```

#Categorical Values
* Name, Cabin, Embarked, Gender

```python
nom_cd=['Name','Ticket','Cabin']#nominal categorical data-no order
ord_cd=['Embarked','Gender']#ordinal categorical data-follow a particular order
```

```python
#find the number of values for gender
data['Gender'].value_counts()
gender_categories=['Female','Male']
#order the categories
data['Gender']=pd.Categorical(data['Gender'],categories=gender_categories,ordered=True)
#display the new ordered gender column
data['Gender']
```

```python
#convert gender from categorical to numerical
data['Gender']=data['Gender'].cat.codes
data['Gender']
```

```python
median_index=data['Gender'].median()
data['Gender'].fillna(median_index,inplace=True)
#display the output
data['Gender']
```
